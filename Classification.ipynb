# **Data Preprocessing and data split**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
y.head()
y[:10]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

# **KNN** **Model**

```python
# Fitting clasifier to the Training set
# Loading libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score

# Instantiate learning model (k = 3)
classifier = KNeighborsClassifier(n_neighbors=3)

# Fitting the model
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
# Evaluating KNN prediction Model

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy of our KNN model is equal ',  accuracy)
```

# **SVM model**

# Tune hyperparmaters



# 1. Regularization (C)

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)

accuracy = model_C.score(X_test, y_test)
print("Accuracy of SVM:", accuracy)
```

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)

accuracy = model_C.score(X_test, y_test)
print("Accuracy of SVM:", accuracy)
```

# 2. Gamma

```python
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)


accuracy = model_g.score(X_test, y_test)
print("Accuracy of SVM:", accuracy)
```

# 3. Kernel (linear, radial basis function (RBF), polynomial, and Sigmoid kernel)

```python
model_linear= SVC(kernel='linear')
model_linear.fit(X_train, y_train)


accuracy = model_linear.score(X_test, y_test)
print("Accuracy of SVM_linear:", accuracy)
```

```python
model_rbf = SVC(kernel='rbf')
model_rbf.fit(X_train, y_train)

accuracy = model_rbf.score(X_test, y_test)
print("Accuracy of SVM_rbf:", accuracy)
```

```python
model_Polynomial = SVC(kernel='poly')
model_Polynomial.fit(X_train, y_train)


accuracy = model_Polynomial.score(X_test, y_test)
print("Accuracy of SVM_poly:", accuracy)
```

```python
model_sigmoid = SVC(kernel='sigmoid')
model_sigmoid.fit(X_train, y_train)


accuracy = model_sigmoid.score(X_test, y_test)
print("Accuracy of SVM_sigmoid:", accuracy)
```

# Also possible to define all parmeters once

```python
# Create a kernel support vector machine model
svm = SVC(kernel='rbf', gamma=0.1, C=10.0)

# Train the model on the training data
svm.fit(X_train, y_train)

# Evaluate the model on the test data
accuracy = svm.score(X_test, y_test)
print('Accuracy:', accuracy)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=20)  #by changing the number of estimators or tree we can imporve the perfromance
rf_model.fit(X_train, y_train)
```

```python

# Evaluate the model on the test data
accuracy = rf_model.score(X_test, y_test)
print('Accuracy:', accuracy)
```

# Confusion matrix

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
